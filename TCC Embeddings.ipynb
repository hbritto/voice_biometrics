```python
from deep_speaker.audio_reader import AudioReader
from deep_speaker.unseen_speakers import inference_embeddings
from deep_speaker.constants import c
# from deep_speaker.utils import parallel_function
```

```python
base_dir = '/run/media/hbritto/Data/Datasets/deep-speaker-data'
# base_dir = '/home/hbritto/code/tcc/deep-speaker-data'
audio_dir = base_dir + '/VCTK-Corpus/'
cache_dir = base_dir + '/cache/'

audio_reader = AudioReader(input_audio_dir=audio_dir, output_cache_dir=cache_dir, sample_rate=c.AUDIO.SAMPLE_RATE, multi_threading=True)
```

```python
from time import time
import pickle
import os
all_embs = dict()
speaker_ids = audio_reader.all_speaker_ids
t1 = time()
# print(parallel_function(inference_embeddings, ['p225'], 8))
for speaker in sorted(speaker_ids):
    emb = inference_embeddings(audio_reader, speaker)
    all_embs[speaker] = emb


with open(os.path.join(cache_dir, 'embeddings.pkl'), 'wb') as f:
    pickle.dump(all_embs, f, protocol=pickle.HIGHEST_PROTOCOL)

    
# emb = inference_embeddings(audio_reader, 'p225')
```
